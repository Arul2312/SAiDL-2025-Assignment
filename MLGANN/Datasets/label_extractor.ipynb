```python
import numpy as np
import pandas as pd
```

```python
dti_df = pd.read_csv("Datasets/Data/dti_dataset.csv")
```

```python
targets = dti_df["PDB ID"].unique()
drugs = dti_df["Drug IDs"].unique()

unique_drugs = []

for drug in drugs:
    # print(drug)
    for unq_drug in drug.split(';'):
        # print(unq_drug.strip())
        unique_drugs.append(unq_drug.strip())
        
        
unique_drugs = np.array(unique_drugs)
```

```python
dti_dict = {}

for _, row in dti_df.iterrows():
    drug_ids = row['Drug IDs'].split('; ')
    target_id = row['PDB ID']
    for drug_id in drug_ids:
        if drug_id not in dti_dict:
            dti_dict[drug_id] = set()
        dti_dict[drug_id].add(target_id)
```

```python
dti_dict
```

```python
interaction_matrix = np.zeros((len(unique_drugs), len(targets)), dtype=int)
interaction_matrix.shape
```

```python
interaction_df = pd.DataFrame(interaction_matrix, index=unique_drugs, columns=targets)
interaction_df
```

```python
for interactions in dti_dict:
    # print(f"{interactions}: {dti_dict[interactions]}")
    for targ in dti_dict[interactions]:
        interaction_df.loc[interactions, targ] = 1

interaction_df.head()
```

```python
interaction_df.to_csv('Datasets/Data/target_labels.csv')
```
